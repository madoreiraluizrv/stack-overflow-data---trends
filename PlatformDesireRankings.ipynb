# Gather and Assess
First, we load survey results from each of the three last years, and then check each dataframe head to assess it

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from collections import defaultdict

df_17 = pd.read_csv('./survey_results_public_2017.csv')
print(df_17.head())

df_18 = pd.read_csv('./survey_results_public_2018.csv')
print(df_18.head())

df_19 = pd.read_csv('./survey_results_public_2019.csv')
print(df_19.head())
```

# Clean and Analyze
Now, we clean the language column, so we can better analyze which are the platform most desired to be used in the last years

```python
def total_count(df, col1, col2, look_for):
    '''
    INPUT:
    df - the pandas dataframe you want to search
    col1 - the column name you want to look through
    col2 - the column you want to count values from
    look_for - a list of strings you want to search for in each row of df[col]

    OUTPUT:
    new_df - a dataframe of each look_for with the count of how often it shows up
    '''
    new_df = defaultdict(int)
    #loop through list of ed types
    for val in look_for:
        #loop through rows
        for idx in range(df.shape[0]):
            #if the ed type is in the row add 1
            if val in df[col1][idx]:
                new_df[val] += int(df[col2][idx])
    new_df = pd.DataFrame(pd.Series(new_df)).reset_index()
    new_df.columns = [col1, col2]
    new_df.sort_values('count', ascending=False, inplace=True)
    return new_df
```

```python
possible_vals = ['Amazon Echo','Amazon Web Services (AWS)','Android','Apple Watch or Apple TV','Arduino','AWS','Azure','Docker',
                 'Drupal','ESP8266','Firebase','Gaming console','Google Cloud Platform','Google Cloud Platform/App Engine',
                 'Google Home','Heroku','IBM Cloud or Watson','iOS','Kubernetes','Linux','Linux Desktop','Mac OS','MacOS',
                 'Mainframe','Microsoft Azure','NA','Other(s):','Predix','Raspberry Pi','Salesforce','Serverless','SharePoint',
                 'Slack','Windows','Windows Desktop','Windows Desktop or Server','Windows Phone','WordPress']

def clean_and_plot(df, column, title='Platform desired', plot=True):
    '''
    INPUT 
        df - a dataframe holding the platform desired column
        title - string the title of your plot
        column - name of the column that contains platform desired responses
        plot - bool providing whether or not you want a plot back
        
    OUTPUT
        platform_desired_df - a dataframe with the percentage of individuals that desire to work with each platform
        Displays a plot of pretty things related to the platform desired column.
    '''
#   Gets the count of how manuy individuals wants to work with each platform
    platform_desired = df[column].value_counts().reset_index()
    platform_desired.rename(columns={'index': 'platform', column: 'count'}, inplace=True)
    platform_desired_df = total_count(platform_desired, 'platform', 'count', possible_vals)

#   Plots the graph with the percentage of individuals that wants to work with each platform
    platform_desired_df.set_index('platform', inplace=True)
    if plot:
        (platform_desired_df/platform_desired_df.sum()).plot(kind='bar', legend=None);
        plt.title(title);
        plt.show()

#   Create the dataframe with the percentage of individuals that wants to work with each platform and returns it
    platform_desired_df = platform_desired_df/platform_desired_df.sum()
    return platform_desired_df
    
platform_desired_df_17 = clean_and_plot(df_17, 'WantWorkPlatform')
platform_desired_df_18 = clean_and_plot(df_18, 'PlatformDesireNextYear')
platform_desired_df_19 = clean_and_plot(df_19, 'PlatformDesireNextYear')
```

# Model and Visualize
Then, we finally make a ranking model, so we can visualize the evolution of each platform between the years

```python
def ranking(df17, df18, df19, plot=True):
    '''
    INPUT 
        df17 - a dataframe holding how many individuals wanted to work with each platform in 2017
        df18 - a dataframe holding how many individuals wanted to work with each platform in 2018
        df19 - a dataframe holding how many individuals wanted to work with each platform in 2019
        plot - bool providing whether or not you want a plot back
        
    OUTPUT
        df - a dataframe with the ranking of each platform, by year
        Displays a plot of with the ranking evolution of each platform between 2017 and 2019.
    '''
#   Rename columns, so we have information for each year when we merge all dataframes
    df17.rename(columns={'platform': 'platform', 'count': '2017'}, inplace=True)
    df18.rename(columns={'platform': 'platform', 'count': '2018'}, inplace=True)
    df19.rename(columns={'platform': 'platform', 'count': '2019'}, inplace=True)

#   Create ranking column
    df17['ranking_2017'] = range(1, len(df17) + 1)
    df18['ranking_2018'] = range(1, len(df18) + 1)
    df19['ranking_2019'] = range(1, len(df19) + 1)

#   Merge dataframes into a unique one, with information from all 3 years
    df17_18 = pd.merge(df17, df18, on='platform')
    df = pd.merge(df17_18, df19, on='platform')

#   Drop percentage columns, mantaining only the ranking ones
    df = df.drop(['2017', '2018', '2019'], 1)

#   Negates all values, so the most used language will appear on top when we plot, then sort based on 2019 ranking
    df = -df
    df = df.sort_values(by=['ranking_2019'], ascending=False)
    
#   Transpose the dataframe
    df = df.T
    
#   Plot the graph showing the ranking evolution of each language between 2017 and 2019
    if (plot):
        df = df.reset_index()
        ax = df.plot()
        ax.set_xticks(df.index)
        ax.set_xticklabels(df['index'], rotation=90)
        ax.axes.get_yaxis().set_visible(False)
        ax.legend(loc=(1.04,0))
    
    return df

platform_desired_ranking = ranking(platform_desired_df_17, platform_desired_df_18, platform_desired_df_19)
```
